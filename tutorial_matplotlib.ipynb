```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.available
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
plt.plot();
```

```python
x = [1,2,3,4]
y = [10,8,6,9]
plt.plot(x,y, c = 'green');
```

```python
#pyplot API
x = np.linspace(0,10,1000)
a = [0,2,4,6,8,10]
b = [0,0,0,0,0,0]
plt.plot(x,np.sin(x),linestyle = "dashed", label = "Sin(x)");
plt.plot(a,b, color = 'red',label = "y=0")
plt.xlabel("Variable x");
plt.ylabel("Sin(x)");
plt.axis("equal")
plt.legend();
```

```python
x = [1,2,3,4]
y = [11,22,33,44]
fig, ax = plt.subplots(figsize = (5,5));
ax.plot(x,y);
ax.set(title = "A simple plot",
       xlabel = "Variable X",
       ylabel = "Variable Y");
```

```python
# graph line 
x = np.linspace(0,10,100)
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

```python
# scatter graph
# 1: using plot API
plt.scatter(x,np.exp(x)); # y = e^(x)
```

```python
# scatter using OO API
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
size =1000*rng.randn(100)
color = rng.randn(100)
fig, ax = plt.subplots()
img = ax.scatter(x,y, s = size, c = color,cmap = 'viridis',alpha = 0.3);
fig.colorbar(img);
```

```python
soft_dink_price = {"coke":10,
                   "Pessi":12,
                   "Sprite": 8
                   }
fig, ax = plt.subplots()
ax.bar(soft_dink_price.keys(), soft_dink_price.values())
ax.set(title = "Soft Dink Price",
       xlabel = "name of soft dink",
       ylabel = "price of soft dink")
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_dink_price.keys()),list(soft_dink_price.values()))
```

```python
# histogram
np.random.seed(42)
student_height = np.random.normal(170,10,250)
plt.hist(student_height, bins = 30)
```
