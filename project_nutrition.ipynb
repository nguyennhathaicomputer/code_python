```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cereal_df = pd.read_csv('nutrition.csv')
cereal_df
```

```python
#irrelevant fields
fields = ['shelf','weight','cups','rating']
cereal_new_df = cereal_df.drop(fields, axis = 1)
cereal_new_df
```

```python
dic = {"calories":cereal_new_df['calories'],
       "protein":cereal_new_df['protein'],
       "fat":cereal_new_df['fat'],
       "sodium":cereal_new_df['sodium'],
       "fiber":cereal_new_df['fiber'],
       "carbo":cereal_new_df['carbo'],
       "sugars":cereal_new_df['sugars'],
       "potass":cereal_new_df['potass'],
       "vitamins":cereal_new_df['vitamins']
       }
temp = pd.DataFrame(dic)

cereal_corr = temp.corr() # get correlation data
cereal_corr
```

```python
#ones_link can build a matrix of boolean with the same shape  as our data
ones_corr = np.ones_like(cereal_corr, dtype = bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
# function triu  in numpy : return only upper triangle matrix
mask = np.triu(ones_corr)
mask
```

```python
test = sns.diverging_palette(145, 300, s=60, as_cmap=True)
sns.heatmap(data = cereal_corr, mask = mask, cmap =test)
```

```python
adjusted_mask = mask[1:,:-1]
adjusted_mask
```

```python
adjusted_ceraeal_corr = cereal_corr.iloc[1:,:-1]
adjusted_ceraeal_corr
```

```python
fig, ax = plt.subplots()
cmap = sns.diverging_palette(0,230,90,60, as_cmap = True)

sns.heatmap(data = adjusted_ceraeal_corr, mask = adjusted_mask, annot= True, fmt = ".2f", cmap = cmap,vmin = -1, vmax =1, linewidths =1)
```
